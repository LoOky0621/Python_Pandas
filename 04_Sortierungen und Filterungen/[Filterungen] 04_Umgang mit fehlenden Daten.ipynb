### Befüllung von leeren Einträgen

#### Numerische Null

```python
umsatz.Umsatz.fillna(0, inplace=True)
```

```python
umsatz
```

#### Arithmetisches Mittel

```python
umsatz.Umsatz.fillna(umsatz.Umsatz.mean(), inplace=True)
```
