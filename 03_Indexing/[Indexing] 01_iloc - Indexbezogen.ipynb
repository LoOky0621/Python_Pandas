```python
umsatz.iloc[:, 11]
```

```python
umsatz.columns
```

```python
umsatz.head(1)
```

```python
umsatz.iloc[:, 5:ncol]
```

```python
umsatz.iloc[:, [0, 4, ncol-1]]
```
